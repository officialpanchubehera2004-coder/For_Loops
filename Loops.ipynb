# Pattern print

```python
for i in range(4):
    for j in range(i+1):
        print("#",end=' ')
    print()  
```

```python
for i in range(6):
    print('# '*i)
```

```python
for i in range(4):
    for j in range(4-i):
        print("#",end=' ')
    print()  
```
